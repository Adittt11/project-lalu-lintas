UBER CROSS

```python
def count_rows(rows):
    return len(rows)

by_cross = df.groupby(["weekday","hour"]).apply(count_rows)
by_cross
```

```python
pivot = by_cross.unstack()
pivot
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(pivot)
```

```python
def heatmap(col1, col2):
    by_cross = df.groupby([col1, col2]).apply(lambda x:len(x))
    pivot = by_cross.unstack()
    plt.figure(figsize=(10,6))
    return sns.heatmap(pivot,annot=False)

heatmap("day","hour")
```

```python
df = df[df["month"]==4]
```

```python
heatmap("weekday","month")
```

```python
plt.figure(figsize=(10,5))

plt.plot(df['Lon'],df["Lat"],"b+",ms=10)
plt.xlim(-74.2, -73.7)
plt.ylim(40.6,41)
```

```python
df_out = df[df["weekday"]== "Sunday"]
df_out.head()
```

```python
from folium.plugins import HeatMap
import folium

basemap = folium.Map()

test = df_out.groupby(["Lat", "Lon"])["weekday"].count().reset_index()
HeatMap(test, zoom = 20, radius=15).add_to(basemap)

basemap

```

```python
def plot_heatmap(df, day):
    df_out = df[df["weekday"]== day]
    test = df_out.groupby(["Lat", "Lon"])["weekday"].count().reset_index()
    HeatMap(test, zoom = 20, radius=15).add_to(basemap)
    return basemap
```

```python
plot_heatmap(df, "monday")
```
